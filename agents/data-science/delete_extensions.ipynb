````python
import requests
````

````python
project = "36231825761"
location = "us-central1"
service_name = "aiplatform.googleapis.com"
version = "v1beta1"

# Get an access token with `gcloud auth print-access-token` and paste below.
token = ""
````

````python
base_url = f"https://{location}-{service_name}/{version}"
````

````python
headers = {
    "Authorization": f"Bearer {token}",
    "Content-Type": "application/json",
}
````

Get a list of extensions

````python
response = requests.get(
    url=f"{base_url}/projects/{project}/locations/{location}/extensions",
    headers=headers,
)
````

````python
response.json()
````

I first used `curl` and `jq` to confirm the response top level keys. It's one key, `extensions`, that contains an array of extensions. Each extension has a top-level `name` key, which is the full resource name of the extension.
```sh
curl -X GET https://$location-$service_name/$version/projects/36231825761/locations/us-central1/extensions -H "Authorization: Bearer $(gcloud auth print-access-token)" | jq 'keys'
```


Filter the response list for the extension names.

````python
extension_names = [
    extension.get("name")
    for extension in response.json().get("extensions", [])
]
extension_names
````

Delete each extension by name.

````python
for extension_name in extension_names:
    response = requests.delete(
        url=f"{base_url}/{extension_name}",
        headers=headers,
    )
    response.raise_for_status()
    if response.status_code == 200:
        print(response.text)
````
